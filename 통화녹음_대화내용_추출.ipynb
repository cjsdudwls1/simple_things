```python
# 1. Gemini 패키지 설치
!pip install google-generativeai

# 2. 라이브러리 불러오기
import google.generativeai as genai
from google.colab import files

from google.colab import userdata
a=userdata.get('Gemini_API_KEY')

genai.configure(api_key=a)
```

```python
# 4. 파일 업로드
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# 5. 모델 불러오기 (멀티모달 지원)
model = genai.GenerativeModel("gemini-2.5-pro")

# 6. 오디오 파일 읽기
import pathlib
audio_file = pathlib.Path(file_name)

# 7. 전사 요청
response = model.generate_content([
    {"mime_type": "audio/m4a", "data": audio_file.read_bytes()},
    "이 오디오 파일을 한국어 텍스트로 전사해줘."
])

# 8. 전사 결과 출력
print(response.text)

# 9. 텍스트 파일로 저장
with open("transcribed_text.txt", "w", encoding="utf-8") as f:
    f.write(response.text)
```
