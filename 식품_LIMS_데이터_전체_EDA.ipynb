```python
# prompt: ['자가품질', '모니터링', '검사명령', '수입검사', '압몰수품', '수거검사', '기타', '제조수입', '품질검사'모두 데이터 프레임으로 로드

import pandas as pd

dataframes = {}
keys_to_load = ['자가품질', '모니터링', '검사명령', '수입검사', '압몰수품', '수거검사', '기타', '제조수입', '품질검사']

for key in keys_to_load:
    if key in loaded_data:
        dataframes[key] = loaded_data[key]
        print(f"\n'{key}' 데이터프레임이 성공적으로 로드되었습니다.")
        print(dataframes[key].head())
    else:
        print(f"\n읽어온 데이터에 '{key}' 키가 없습니다.")
```

```python
for key, df in dataframes.items():
    print(f"\n--- Analysis for '{key}' ---")

    # Filter for '부적합' data
    inadequate_df = df[df['EXPR_IEM_JDGMNT_NAME'] == '부적합']

    total_count = len(df)
    inadequate_count = len(inadequate_df)

    if total_count > 0:
        inadequate_percentage = (inadequate_count / total_count) * 100
        print(f"\nPercentage of '부적합' data for '{key}': {inadequate_percentage:.2f}% ({inadequate_count}/{total_count})")
    else:
        print(f"\nNo data found for '{key}'.")


    if not inadequate_df.empty:
        print(f"\nValue counts for 'INSPCT_KND_NAME' in '부적합' data for '{key}':")
        display(inadequate_df['INSPCT_KND_NAME'].value_counts().reset_index(name='Count').rename(columns={'index': 'INSPCT_KND_NAME'}))
    else:
        print(f"\nNo '부적합' data found for '{key}'.")

    # Display all unique 'PRDUCT_NM' in the original dataframe
    print(f"\nUnique 'PRDUCT_NM' in the original data for '{key}':")
    display(df['PRDUCT_NM'].unique())
```

```python
for key, df in dataframes.items():
    print(f"\n--- Value Counts for 'PRDUCT_NM' in '{key}' ---")
    if 'PRDUCT_NM' in df.columns:
        display(df['PRDUCT_NM'].value_counts().reset_index(name='Count').rename(columns={'index': 'PRDUCT_NM'}))
    else:
        print(f"'{key}' dataframe does not have a 'PRDUCT_NM' column.")
```
